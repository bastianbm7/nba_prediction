```python
import os
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Define the folder path where your Excel files are located
folder_path = 'data'

# Get a list of all Excel files in the folder
excel_files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx')]

# Create an empty list to store DataFrames
dfs = []

# Loop through the Excel files, read them into DataFrames, and append them to the list
for file in excel_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_excel(file_path)  # Use pd.read_excel() for Excel files
    dfs.append(df)

# Concatenate all DataFrames in the list into one DataFrame
concatenated_df = pd.concat(dfs, ignore_index=True)
concatenated_df
```

```python
concatenated_df = concatenated_df.loc[concatenated_df['PERIOD']<= 4 ]
total_points_by_period = concatenated_df.groupby(['GAME_ID', 'Season', 'PERIOD'])['POINTS'].sum().reset_index()

# Now, create the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='PERIOD', y='POINTS', hue='Season', data=total_points_by_period)

# Set plot title and labels
plt.title('Boxplot of Total Points by Period')
plt.xlabel('Period')
plt.ylabel('Total Points')

# Show the plot
plt.show()
```
